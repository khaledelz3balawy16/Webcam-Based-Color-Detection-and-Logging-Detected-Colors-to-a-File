```python
import cv2
import numpy as np
import webcolors

# قائمة الألوان القياسية مع قيم RGB
color_names = {
    "BLACK": (0, 0, 0),
    "WHITE": (255, 255, 255),
    "RED": (255, 0, 0),
    "LIME": (0, 255, 0),
    "BLUE": (0, 0, 255),
    "YELLOW": (255, 255, 0),
    "CYAN": (0, 255, 255),
    "MAGENTA": (255, 0, 255),
    "SILVER": (192, 192, 192),
    "GRAY": (128, 128, 128),
    "MAROON": (128, 0, 0),
    "OLIVE": (128, 128, 0),
    "GREEN": (0, 128, 0),
    "PURPLE": (128, 0, 128),
    "TEAL": (0, 128, 128),
    "NAVY": (0, 0, 128)
}

# دالة لحساب أقرب لون باستخدام المسافة الإقليدية بين الألوان
def get_closest_color_name(rgb_color):
    min_distance = float('inf')
    closest_color_name = "Undefined"

    # المرور عبر كل الألوان في القائمة وتحويلها إلى RGB
    for color_name, color_rgb in color_names.items():
        distance = np.sqrt(np.sum((np.array(color_rgb) - np.array(rgb_color)) ** 2))
        if distance < min_distance:
            min_distance = distance
            closest_color_name = color_name
    return closest_color_name

# فتح الكاميرا وضبط الدقة
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# فتح ملف لتسجيل الألوان
with open("detected_colors.txt", "w") as color_file:

    while True:
        _, frame = cap.read()

        # عكس الصورة لتكون متوافقة مع حركة المستخدم
        frame = cv2.flip(frame, 1)

        # تطبيق Gaussian Blur لتقليل الضوضاء وتحسين دقة اللون
        frame = cv2.GaussianBlur(frame, (7, 7), 0)

        # تحويل الصورة من BGR إلى HSV
        hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        height, width, _ = frame.shape

        # الحصول على متوسط لون المنطقة المحيطة بالبكسل المركزي
        cx, cy = int(width / 2), int(height / 2)
        region_size = 15
        region = hsv_frame[cy - region_size:cy + region_size, cx - region_size:cx + region_size]
        avg_color_hsv = region.mean(axis=0).mean(axis=0)
        
        # تحويل اللون المتوسط من HSV إلى BGR لعرضه كخلفية
        avg_color_bgr = cv2.cvtColor(np.uint8([[avg_color_hsv]]), cv2.COLOR_HSV2BGR)[0][0]
        b, g, r = int(avg_color_bgr[0]), int(avg_color_bgr[1]), int(avg_color_bgr[2])

        # تحويل اللون من BGR إلى RGB لتوافق المكتبة
        rgb_color = (r, g, b)

        # الحصول على اسم اللون بناءً على قيم RGB
        color_name = get_closest_color_name(rgb_color)

        # عرض اللون في مستطيل بدون النص
        cv2.rectangle(frame, (cx - 220, 10), (cx + 200, 120), (b, g, r), -1)

        # تسجيل اسم اللون في الملف
        color_file.write(f"{color_name}: RGB({r}, {g}, {b})\n")

        # عرض الصورة
        cv2.circle(frame, (cx, cy), 5, (25, 25, 25), 3)  # وضع دائرة لتحديد المركز
        cv2.imshow("Frame", frame)

        # الخروج عند الضغط على مفتاح "ESC"
        key = cv2.waitKey(1)
        if key == 27:
            break

# تحرير الكاميرا وإغلاق النوافذ
cap.release()
cv2.destroyAllWindows()
```
